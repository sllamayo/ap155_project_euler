```python
import numpy as np
import matplotlib.pyplot as plt

#Defining the Function T(N)

def T(N):
  length = 0
  while N != 1:
    if N % 2 == 0:
      N //= 2
    else: N = 3 * N + 1
    length += 1
  return length

#Vectorizing the output
vectorizedcollatz = np.vectorize(T)

Nvalues = range(1,10001)

collatzfunction = vectorizedcollatz(Nvalues)

#Plotting the collatz sequence
plt.scatter(Nvalues, collatzfunction, s = 10)



```
